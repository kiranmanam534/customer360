```python
base_adls2_path = "abfss://customer360unitycatalog@kmstorage9490.dfs.core.windows.net"
```

```python
df = spark.readStream.format("cloudFiles") \
    .option("cloudFiles.format", "csv") \
    .option("cloudFiles.schemaLocation", f"{base_adls2_path}/schema/orders") \
    .option("cloudFiles.maxFilesPerTrigger", 10) \
    .load(f"{base_adls2_path}/raw/orders")
```

```python

df.writeStream \
    .format("parquet") \
    .outputMode('append') \
    .option("checkpointLocation", f"{base_adls2_path}/checkpoints/orders") \
    .option("path", f"{base_adls2_path}/bronze/orders") \
    .trigger(once=True) \
    .toTable("customer360_cata.bronze.orders").awaitTermination()
```
