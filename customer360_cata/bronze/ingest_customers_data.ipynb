```python
base_adls2_path = "abfss://customer360unitycatalog@kmstorage9490.dfs.core.windows.net"
```

```python
df = spark.readStream.format("cloudFiles") \
    .option("cloudFiles.format", "csv") \
    .option("cloudFiles.schemaLocation", f"{base_adls2_path}/schema/customers") \
    .option("cloudFiles.schemaHints", "customer_id string, name string, dob string, gender string, location_id int, EMAIL string") \
    .option("cloudFiles.maxFilesPerTrigger", 10) \
    .load(f"{base_adls2_path}/raw/customers")
```

```python

df.writeStream \
    .format("parquet") \
    .outputMode('append') \
    .option("checkpointLocation", f"{base_adls2_path}/checkpoints/customers") \
    .option("path", f"{base_adls2_path}/bronze/customers") \
    .trigger(once=True) \
    .toTable("customer360_cata.bronze.customers").awaitTermination()
```
